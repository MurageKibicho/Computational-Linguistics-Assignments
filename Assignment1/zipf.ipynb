```python
import nltk
import string
from nltk import FreqDist
from collections import OrderedDict
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
nltk.download('gutenberg')
```

```python
gutenbergFrequencyDistribution = FreqDist(nltk.corpus.gutenberg.words())
sortedFrequencyDictionary = OrderedDict(sorted(gutenbergFrequencyDistribution.items(),reverse=True, key=lambda t: t[1]))
gutenbergWordsRank = list(sortedFrequencyDictionary.keys())
gutenbergFrequenciesList = list(sortedFrequencyDictionary.values())
length = len(gutenbergFrequenciesList)
plt.loglog(range(1, length + 1),gutenbergFrequenciesList)
            
plt.ylabel("frequency(f)",fontsize=14,fontweight='bold')
plt.xlabel("rank(r)",fontsize=14,fontweight='bold')
plt.grid(True)


```

```python
f = open('europarl-v7.ro-en.en')
englishText = f.readlines()
englishTextFrequencyDistribution = FreqDist(englishText)
sortedEnglishFrequencyDictionary = OrderedDict(sorted(englishTextFrequencyDistribution.items(),reverse=True, key=lambda t: t[1]))
englishWordsRank = list(sortedEnglishFrequencyDictionary.keys())
englishFrequenciesList = list(sortedEnglishFrequencyDictionary.values())
englishLength = len(englishFrequenciesList)
plt.loglog(range(1, englishLength + 1),englishFrequenciesList)

plt.ylabel("frequency(f)",fontsize=14,fontweight='bold')
plt.xlabel("rank(r)",fontsize=14,fontweight='bold')
plt.grid(True)
```

```python
f = open('europarl-v7.ro-en.ro')
russianText = f.readlines()
russianTextFrequencyDistribution = FreqDist(russianText)
sortedRussianFrequencyDictionary = OrderedDict(sorted(russianTextFrequencyDistribution.items(),reverse=True, key=lambda t: t[1]))
russianWordsRank = list(sortedRussianFrequencyDictionary.keys())
russianFrequenciesList = list(sortedRussianFrequencyDictionary.values())
russianLength = len(russianFrequenciesList)
plt.loglog(range(1, russianLength + 1),russianFrequenciesList)

plt.ylabel("frequency(f)",fontsize=14,fontweight='bold')
plt.xlabel("rank(r)",fontsize=14,fontweight='bold')
plt.grid(True)
```

1. Yes, the same zipfian distribution is observed
2. A difference I observe is in the shape of the top-most left grid, the one bounded by the frequency range(10^2 - 10^3). In the Englist text,it forms more of a straight line compared to the russian text. I believe this is mostly caused by differences in punctuation and grammatical styles.
